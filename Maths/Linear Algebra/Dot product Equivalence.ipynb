#### Two definitions

Imagine two vectors $ A =[a_x,a_y]$ and $ B = [b_x,b_y]  $

The dot product is given by following two definitions:  

1. $ a_xb_x + a_yb_y $
2. $|A||B|cos \theta $  

#### How these two definitions are equivalent?  

One proof based on components is here : https://math.stackexchange.com/a/2629588

##### Alternatively:

![dot_product_equivalence](dot_product_equivalence_2.png)

(Source: https://www.geogebra.org/calculator/artwuzxc)

Imagine a plane with x and y axis. Let's say vector A makes angle of $\theta_1$ with the x-axis, and vector B makes angle of $\theta_2 > \theta_1$ with x-axis, and let $\theta$ be the angle between the two vectors.

Therefore,

$  a_x = |A|cos\theta_1$  
$  a_y = |A|sin\theta_1$  
$  b_x = |B|cos\theta_2$  
$  b_y = |B|sin\theta_2$  
$ \theta = \theta_2 - \theta_1 $  

$ a_xb_x + a_yb_y = |A|cos\theta_1 |B|cos\theta_2 + |A|sin\theta_1|B|sin\theta_2$  

$ = |A||B|cos\theta_1cos\theta_2 + |A||B|sin\theta_1sin\theta_2 $ 

$ = |A||B|(cos\theta_1cos\theta_2 + sin\theta_1sin\theta_2) $ 

![cos_a_minus_b](cos_a_minus_b.png)

In the above figure,$\theta_1$ and $\theta_2$ are any general angles (not to be correlated with the angles in the first figure).  $\angle AFE $ and $\angle AEB $ are chosen to be right angles.


$\angle EAB + \angle ABD + \angle DBE = 90^o$  

$\theta_2 - \theta_1 + \angle ABD + \angle DBE = 90^o$

$\theta_2 + \angle ABD - \theta_1 + \angle DBE = 90^o$$\tag 1$  


$\angle CAE + \angle EAB + \angle ABD = 90^o$  

$ \theta_1 + (\theta_2 - \theta_1) + \angle ABD =  90^o$  

$ \theta_2 + \angle ABD =  90^o$$\tag 2$  


Substituting (2) in (1)  

$90^o - \theta_1 + \angle DBE = 90^o$   

$ - \theta_1 + \angle DBE = 0$   

$ \angle DBE = \theta_1 = \angle CAE $$\tag 3$   

Therefore,  

$tan\theta_1 = \frac{EF}{AF} = \frac{DE}{BD} = \frac{CF}{BD} $  

$ BD*EF = AF*CF $$\tag 4$




$ cos\theta_1cos\theta_2 + sin\theta_1sin\theta_2$  

$ = \frac{AC}{AB} \frac{AF}{AE} + \frac{BC}{AB} \frac{EF}{AE}$  

$ = \frac{AC*AF+BC*EF}{AB*AE}$  

$ = \frac{(AF-CF)AF+(BD+EF)EF}{AB*AE}$  

$ = \frac{AF^2 - AF*CF + BD*EF +EF^2}{AB*AE}$  

$ = \frac{AF^2 +EF^2 + BD*EF - AF*CF }{AB*AE}$  

$ = \frac{AE^2 + BD*EF - AF*CF }{AB*AE}$  

$ = \frac{AE^2 + 0 }{AB*AE}$$\tag{from 4}$  

$ = \frac{AE}{AB}$  

$ = cos(\angle BAE)$  

$ = cos(\theta_2 - \theta_1)$  

$ = cos\theta $  

Therefore in general:  

$ cos\theta_1cos\theta_2 + sin\theta_1sin\theta_2 = cos(\theta_2 - \theta_1) = cos\theta$  

Now returning to our original equation:  

$ a_xb_x + a_yb_y = |A||B|(cos\theta_1cos\theta_2 + sin\theta_1sin\theta_2)$  

$ = |A||B|cos(\theta_2 - \theta_1) $ 

$ = |A||B|cos(\theta) $ 


#### Alternate Proof of : $ cos(\theta_2 - \theta_1) $  


$cos\theta$ and $sin\theta$ are basically respectively x and y co-ordinates of the intersection of the radius at angle $\theta$ with x-axis and the unit circle. When the angle is $-\theta$, x-co-ordinate remains same and positive while the y-co-ordinate becomes negative.  

$ \therefore x = cos(-\theta) = cos(\theta)$ and $ y = sin(-\theta)=-sin(\theta) $  

Proof of $cos(\theta_1 + \theta_2)$: https://www.youtube.com/watch?v=V3-xCPDzQ1Q  


$ cos(\theta) = cos(\theta_2 - \theta_1) $  

$ = cos(\theta_2 + (-\theta_1)) $  

$ = cos\theta_2 cos(-\theta_1) - sin(\theta_2)sin(-\theta_1)$  

$ = cos\theta_2 cos\theta_1 + sin\theta_2sin\theta_1$
