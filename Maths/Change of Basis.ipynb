## Change of basis


Change of basis is very much linked to the linear transformation we had seen earlier.  

Different people might have different perception about the same vector. The same vector can be represented in many different co-ordinate systems with different co-ordinates.

If we transform our original basis vectors (along with any other vectors) to the changed basis vectors, i.e. apply the tranformation containing the basis vectors in the transformation matrix as the target basis vectors (from source basis vectors perspective), we would get the co-ordinates of the original basis vectors and any other transformed vectors from the point of view of the target basis vectors.  

This is kind of relativity. Like when we are at rest and observe speed of a moving vehicle A as 40 km/hr.  
If we get into another vehicle B and follow A with 20 km/hr, we'll observe the relative speed of the A is 20 km/hr.  
If increase the speed of B to 40 km/hr; we'll observe A to be static.  
If we further increase the speed of B to 60 km/hr and overtake A, we'll observe A to be going back with 20 km/hr.  

If we want to know the relative speed of all vehicles from B when it is at x km/hr, we need to subtract x from the speed of all vehicles as observed when B was at rest. We are changing our basis from 0 to x, i.e. transforming the basis speed from 0 to x.  

The vice versa is also applicable.  

If we observe speeds of all vehicles when B is at x km/hr, and we want to know the speeds when B would be at rest, we have to add x to speeds of all vehicles as we observed from B when B was at x km/hr. This is we are doing inverse of previous transformation to compensate it. -x by +x.  

The same is applicable in case of change of basis, if we want to know the co-ordinates of a vector in the original system, and we know the co-ordinates of the same vector in the changed system, we need to apply the inverse transformation (to compensate the original transformation).  




How can I know what $\begin{bmatrix} 2 \\ 1 \end{bmatrix}$ in their system is called in our system? If we know their basis vectors, respective to our system, say $\begin{bmatrix} 1 \\ 1 \end{bmatrix}$ and $\begin{bmatrix} 1 \\ 2 \end{bmatrix}$, then $\begin{bmatrix} 2 \\ 1 \end{bmatrix}$ in their system  is $ 2 * \begin{bmatrix} 1 \\ 1 \end{bmatrix} + 1 * \begin{bmatrix} 1 \\ 2 \end{bmatrix}= \begin{bmatrix} 3 \\ 4 \end{bmatrix}$ in our system   

$  \begin{bmatrix} 1 \ 1 \\ 1 \ 2 \end{bmatrix}  \begin{bmatrix} 2 \\ 1 \end{bmatrix}= \begin{bmatrix} 3 \\ 4 \end{bmatrix}$ $\tag 5$  

If $\begin{bmatrix} 2 \\ 1 \end{bmatrix}$ in their system is called as $\begin{bmatrix} 3 \\ 4 \end{bmatrix}$ in our system. Then in other words, $\begin{bmatrix} 3 \\ 4 \end{bmatrix}$ in our system is called as $\begin{bmatrix} 2 \\ 1 \end{bmatrix}$ in their system.  

Now I want to know what $\begin{bmatrix} 3 \\ 4 \end{bmatrix}$ in our system is called in their system (Which we already know). This time we need to represent our basis vectors respective to their system. Say, they are $\begin{bmatrix} a \\ b \end{bmatrix}$ and $\begin{bmatrix} c \\ d \end{bmatrix}$, then the conversion would look like :    


$ 3 * \begin{bmatrix} a \\ b \end{bmatrix} + 4 * \begin{bmatrix} c \\ d \end{bmatrix}= \begin{bmatrix} 2 \\ 1 \end{bmatrix}$  


$  \begin{bmatrix} a \ c \\ b \ d \end{bmatrix}  \begin{bmatrix} 3 \\ 4 \end{bmatrix}= \begin{bmatrix} 2 \\ 1 \end{bmatrix}$  

This looks to be inverse of (5) above.  

$  \begin{bmatrix} a \ c \\ b \ d \end{bmatrix} = \begin{bmatrix} 1 \ 1 \\ 1 \ 2 \end{bmatrix}^{-1} = \begin{bmatrix} +2 \ -1 \\ -1 \ +1 \end{bmatrix} $  


### Relation between linear transformation and change of basis  

![tn_cb](transformation_change_of_basis.png)

Consider that there are two co-ordinate systems A and B in above graph.

All the vectors co-exist in both the systems. Just they are represented with different co-ordinates respective to the basis vectors of that system. Change of basis enables the translation between the two systems. For example in above graph, the vector A34B21 is represented as $\begin{bmatrix} 3 \\ 4 \end{bmatrix}$ in system A while as $\begin{bmatrix} 2 \\ 1 \end{bmatrix}$ in system B. The same is the case with other two vectors though not explicitly named.  

Matrix with target system's basis vectors(respective to our system) is a mechanism to translate the vector represented in the target system into our system.  

$  \begin{bmatrix} 1 \ 1 \\ 1 \ 2 \end{bmatrix}  \begin{bmatrix} 2 \\ 1 \end{bmatrix}= \begin{bmatrix} 3 \\ 4 \end{bmatrix}$$\tag{B to A}$  

Inverse matrix with our system's basis vectors(respective to the target system) is a mechanism to translate the vector represented in our system into the target system.  

$  \begin{bmatrix} +2 \ -1 \\ -1 \ +1 \end{bmatrix}  \begin{bmatrix} 3 \\ 4 \end{bmatrix}= \begin{bmatrix} 2 \\ 1 \end{bmatrix}$$\tag{A to B}$   

The vector A34B21 is represented as $\begin{bmatrix} 3 \\ 4 \end{bmatrix}$ in system A and as $\begin{bmatrix} 2 \\ 1 \end{bmatrix}$ in system B.  

We can draw a vector A21 represented as $\begin{bmatrix} 2 \\ 1 \end{bmatrix}$ in system A also. Now A34B21 is the transformation of A21, i.e. how A21 would look like in the transformed system (After shifting the basis vectors of system A to the basis vectors of system B).  

We can draw a vector B34 represented as $\begin{bmatrix} 3 \\ 4 \end{bmatrix}$ in system B also. Now A34B21 is the transformation of B34, i.e. how B34 would look like in the transformed system (After shifting the basis vectors of system B to the basis vectors of system A).  

**Note that:**   

A34B21 can be named as A34 or B21 from system A and B perspective respectively, but it is the same vector.  

the transformation $  \begin{bmatrix} 1 \ 1 \\ 1 \ 2 \end{bmatrix}$ moves $\begin{bmatrix} 2 \\ 1 \end{bmatrix}$ to  $\begin{bmatrix} 3 \\ 4 \end{bmatrix}$, in both the systems; A21 to A34 in A while B21 to B34 in B. Or differently said, A21 to B21 and A34 to B34.  

the transformation $  \begin{bmatrix} +2 \ -1 \\ -1 \ +1 \end{bmatrix}$ moves $\begin{bmatrix} 3 \\ 4 \end{bmatrix}$ to  $\begin{bmatrix} 2 \\ 1 \end{bmatrix}$, in both the systems; A34 to A21 in A while B34 to B21 in B. Or differently said, B34 to A34 and B21 to A21.  

If we just look at A34B21, it has two names in two different systems A34 or B21. By changing the basis (A to B or B to A), we change our perspective of looking at it.  

If we look at A21 and B34, they get transformed to A34B21, but the transformed co-ordinates are different if we look at them from different systems.  

A matrix $  \begin{bmatrix} a \ c \\ b \ d \end{bmatrix}$ is a transformation when looked from within the same system (A or B), the co-ordinates of the vector changes, or we say the vector gets transformed, but the system reamains the same.  
The same matrix $  \begin{bmatrix} a \ c \\ b \ d \end{bmatrix}$ is change of basis when looked across the systems (A and B), the co-ordinates of the vector remain the same, but the system changes (say, A to B).  
The inverse matrix just moves back the transformed vector to the original vector, OR change the basis to the original system (B to A).  


#### Transformation after change of basis

If we want to apply a transformation $ \begin{bmatrix} 3 \ 1 \\ 2 \ 2 \end{bmatrix} $ (represented in our basis A) to a vector represented in another basis B $ \begin{bmatrix} 1 \ 1 \\ 1 \ 2 \end{bmatrix} $. We can first get the corresponding vector translated into our basis A, then apply the desired transformation and then again translate the resultant vector(which is represented in our basis) back into the other basis B.  

$V_b =  \begin{bmatrix} 2 \\ 1 \end{bmatrix} $ be the vector in another basis B.  

Which is represented in our basis as:  

$V_a = \begin{bmatrix} 1 \ 1 \\ 1 \ 2 \end{bmatrix}\begin{bmatrix} 2 \\ 1 \end{bmatrix} = \begin{bmatrix} 3 \\ 4 \end{bmatrix} $  

After applying the transformation t = $ \begin{bmatrix} 3 \ 1 \\ 2 \ 2 \end{bmatrix} $  

$V_a^t =  \begin{bmatrix} 3 \ 1 \\ 2 \ 2 \end{bmatrix}\begin{bmatrix} 3 \\ 4 \end{bmatrix} = \begin{bmatrix} 13 \\ 14 \end{bmatrix} $  

Now to represent $V_a^t$ in the basis B, we need to multiply it with the inverse of $ \begin{bmatrix} 1 \ 1 \\ 1 \ 2 \end{bmatrix} $  

$V_b^t =  \begin{bmatrix} {+1} \ -1 \\ +1 \ +1 \end{bmatrix} \begin{bmatrix} 13 \\ 14 \end{bmatrix} = \begin{bmatrix} -1 \\ 27 \end{bmatrix}$  

What we did sequencially can be represented as:  

$ \begin{bmatrix} {+1} \ -1 \\ +1 \ +1 \end{bmatrix} \begin{bmatrix} 3 \ 1 \\ 2 \ 2 \end{bmatrix} \begin{bmatrix} 1 \ 1 \\ 1 \ 2 \end{bmatrix}\begin{bmatrix} 2 \\ 1 \end{bmatrix} $  

$ = \begin{bmatrix} {+1} \ -1 \\ +1 \ +1 \end{bmatrix} \begin{bmatrix} 4 \ 5 \\ 4 \ 6 \end{bmatrix} \begin{bmatrix} 2 \\ 1 \end{bmatrix} $  

$ = \begin{bmatrix} +0 \ -1 \\ +8 \ 11 \end{bmatrix} \begin{bmatrix} 2 \\ 1 \end{bmatrix} $  

$ = \begin{bmatrix} -1 \\ 27 \end{bmatrix} $  

The matrix  $\begin{bmatrix} +0 \ -1 \\ +8 \ 11 \end{bmatrix}$ captures the complete process and can be applied directly onto any vector in basis B to get the resultant transformation in the same basis B.  

**In general, a transformation M represented in basis A, can be represented as the transformation $B^{-1}M{B}$ in basis B, where B represents the basis vectors of B.**
