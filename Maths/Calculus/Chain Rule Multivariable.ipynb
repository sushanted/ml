### Chain Rule - Multivariable 

Consider following functions:

$ y = f_1(x)$  
$ z = f_2(x)$  
$ w = f(y,z)$  

![multivariable_diamond](multivariable_diamond.png)
Source: https://www.geogebra.org/calculator/qf97svu6

For simplicity lets consider that w takes some simple form as :  

$w = y^z + yz + y + z $  

Note that this assumption is based on that in general a multivariable function is a function of functions of the variables. For simplicity only involving products,sum or power of those functions.  

y and z could be any complex functions of x.  

Follows the partial derivatives of w with respect to y and z respectively.  

$\frac{\partial{w}}{\partial{y}} = \frac{\partial}{\partial{y}} y^z + \frac{\partial}{\partial{y}} yz + \frac{\partial}{\partial{y}} y + \frac{\partial}{\partial{y}} z$  

$= \frac{\partial}{\partial{y}} y^z + \frac{\partial}{\partial{y}} yz + 1 + 0$  

$= \frac{\partial}{\partial{y}} y^z + z + 1$  


$= y^z (\frac{z}{y}) + z + 1$$\tag 1$  

$\frac{\partial{w}}{\partial{z}} = \frac{\partial}{\partial{z}} y^z + \frac{\partial}{\partial{z}} yz + \frac{\partial}{\partial{z}} y + \frac{\partial}{\partial{z}} z$  

$ = \frac{\partial}{\partial{z}} y^z + y + 0 + 1$  

$ =  y^z \ln(y) + y + 1$$\tag 2$    

Remember y and z both are functions of x, therefore w being function of y and z, is also a function of x. Partial derivative of w with respect to x is therefore the full derivative. We'll use partial derivative here.  

$\frac{\partial{w}}{\partial{x}} = \frac{\partial{}}{\partial{x}} (y^z + yz + y + z)$  

$\frac{\partial{w}}{\partial{x}} = \frac{\partial{}}{\partial{x}} y^z + \frac{\partial{}}{\partial{x}}yz + \frac{\partial{}}{\partial{x}}y + \frac{\partial{}}{\partial{x}}z$  

$\frac{\partial{w}}{\partial{x}} = \frac{\partial{}}{\partial{x}} y^z + \frac{\partial{}}{\partial{x}}yz + \frac{\partial{y}}{\partial{x}} + \frac{\partial{z}}{\partial{x}}$  

Using product rule:  


$\frac{\partial{w}}{\partial{x}} = \frac{\partial{}}{\partial{x}} y^z + y\frac{\partial{}}{\partial{x}}z + z\frac{\partial{}}{\partial{x}}y + \frac{\partial{y}}{\partial{x}} + \frac{\partial{z}}{\partial{x}}$  

$\frac{\partial{w}}{\partial{x}} = \frac{\partial{}}{\partial{x}} y^z + y\frac{\partial{z}}{\partial{x}} + z\frac{\partial{y}}{\partial{x}} + \frac{\partial{y}}{\partial{x}} + \frac{\partial{z}}{\partial{x}}$  

$y^z = e^{z\ln(y)}$

$\frac{\partial{w}}{\partial{x}} = \frac{\partial{}}{\partial{x}} e^{z \ln(y)} + y\frac{\partial{z}}{\partial{x}} + z\frac{\partial{y}}{\partial{x}} + \frac{\partial{y}}{\partial{x}} + \frac{\partial{z}}{\partial{x}}$  

$\frac{\partial{w}}{\partial{x}} = e^{z \ln(y)}\frac{\partial{}}{\partial{x}} z \ln(y) + y\frac{\partial{z}}{\partial{x}} + z\frac{\partial{y}}{\partial{x}} + \frac{\partial{y}}{\partial{x}} + \frac{\partial{z}}{\partial{x}}$  

Again Using product rule:  

$\frac{\partial{w}}{\partial{x}} = e^{z \ln(y)} (z \frac{\partial{}}{\partial{x}}\ln(y) + \ln(y)\frac{\partial{}}{\partial{x}}z) + y\frac{\partial{z}}{\partial{x}} + z\frac{\partial{y}}{\partial{x}} + \frac{\partial{y}}{\partial{x}} + \frac{\partial{z}}{\partial{x}}$  

$\frac{\partial{w}}{\partial{x}} = e^{z \ln(y)} (\frac{z}{y} \frac{\partial{y}}{\partial{x}} + \ln(y)\frac{\partial{z}}{\partial{x}}) + y\frac{\partial{z}}{\partial{x}} + z\frac{\partial{y}}{\partial{x}} + \frac{\partial{y}}{\partial{x}} + \frac{\partial{z}}{\partial{x}}$  

$\frac{\partial{w}}{\partial{x}} = y^z (\frac{z}{y} \frac{\partial{y}}{\partial{x}} + \ln(y)\frac{\partial{z}}{\partial{x}}) + y\frac{\partial{z}}{\partial{x}} + z\frac{\partial{y}}{\partial{x}} + \frac{\partial{y}}{\partial{x}} + \frac{\partial{z}}{\partial{x}}$  

Collecting same suffix terms together:   

$\frac{\partial{w}}{\partial{x}} = y^z \frac{z}{y} \frac{\partial{y}}{\partial{x}}  +  z\frac{\partial{y}}{\partial{x}} + \frac{\partial{y}}{\partial{x}} + y^z\ln(y)\frac{\partial{z}}{\partial{x}} + y\frac{\partial{z}}{\partial{x}} + \frac{\partial{z}}{\partial{x}}$  


$\frac{\partial{w}}{\partial{x}} = (y^z \frac{z}{y}   +  z + 1)\frac{\partial{y}}{\partial{x}} + (y^z\ln(y) + y + 1)\frac{\partial{z}}{\partial{x}}$  

Substituting results from (1) and (2) above:  

$\frac{\partial{w}}{\partial{x}} = \frac{\partial{w}}{\partial{y}}\frac{\partial{y}}{\partial{x}} + \frac{\partial{w}}{\partial{z}}\frac{\partial{z}}{\partial{x}}$  












#### Logic behind above magic

$ w = f(y)g(z)$  

When we differentiate w.r.t. y or z which are actually involved in the above product  

$ \frac{\partial{w}}{\partial{y}} = f(y)\frac{\partial}{\partial{y}}g(z)  + g(z)\frac{\partial}{\partial{y}} f(y)$  

$ \frac{\partial{w}}{\partial{y}} = 0 + g(z)\frac{\partial}{\partial{y}} f(y) + 0 $  

$ \frac{\partial{w}}{\partial{y}} = g(z)f'(y) $$\tag{2.1}$  

$ \frac{\partial{w}}{\partial{z}} = f(y)\frac{\partial}{\partial{z}} g(z) + g(z)\frac{\partial}{\partial{z}} f(y)$  

$ \frac{\partial{w}}{\partial{z}} = f(y)\frac{\partial}{\partial{z}} g(z) + 0 $  

$ \frac{\partial{w}}{\partial{z}} = f(y)g'(z) $$\tag{2.2}$  

As seen above, while applying the product rule the derivative of other term involving another variable becomes 0 and the final derivative is just the product derivative of the function of current variable and the other function.  

When we differentiate w.r.t x (where y and z are functions of x), which is not directly involved in the product  

$ \frac{\partial{w}}{\partial{x}} = f(y)\frac{\partial}{\partial{x}} g(z) + g(z)\frac{\partial}{\partial{x}} f(y)$  

$ \frac{\partial{w}}{\partial{x}} = f(y)g'(z)\frac{\partial{z}}{\partial{x}} + g(z)f'(y)\frac{\partial{y}}{\partial{x}}$  

Here while applying the product rule, again it is product of one function and derivative of another ($f(y)g'(z)$) plus it's symmetry($g(z)f'(y)$). Both the functions get involved (and no term becomes 0). Each of the product is also augmented(by chain rule) with the derivative w.r.t x ($\frac{\partial{z}}{\partial{x}}$ and $\frac{\partial{y}}{\partial{x}}$).    

Substituting 2.1 and 2.2  

$ \frac{\partial{w}}{\partial{x}} = \frac{\partial{w}}{\partial{z}}\frac{\partial{z}}{\partial{x}} + \frac{\partial{w}}{\partial{y}}\frac{\partial{y}}{\partial{x}}$  

Also note that f' and g' are generic derivatives (they are not specific to any variable x,y or z), any further augmentation due to chain rule can be found after them, like $\frac{\partial{z}}{\partial{x}}$ in $f(y)g'(z)\frac{\partial{z}}{\partial{x}}$  







##### More complex functions

$y^z$ types can be represented as products (or functions of the product) as $ e^{z \ln(y)} $, so the above logic is also applicable for exponential relationships between y and z.   

![multivariable_power](multivariable_power.png)

(source: https://www.geogebra.org/m/jmymgzkv)  


Above diagram represents the $y^z$ or $ v = e^{z \ln(y)}$ function represented as a graph.  

$v = e^w $  

By chain rule:  

$ \frac{\partial{v}}{\partial{x}} = e^w \frac{\partial{w}}{\partial{x}} $  

$ \frac{\partial{v}}{\partial{x}} = e^{z \ln{y}} \frac{\partial{w}}{\partial{x}} $  

Again w can be represented in the $ w = f(y)g(z)$ form, where f(y) = $\ln(y)$ and g(z) = z   

And finally we can calculate $\frac{\partial{w}}{\partial{x}} =  \frac{\partial{w}}{\partial{y}}\frac{\partial{y}}{\partial{x}} + \frac{\partial{w}}{\partial{z}}\frac{\partial{z}}{\partial{x}}$  








##### Long chain

![multivariable_long_chain](multivariable_long_chain.png)

(source: https://www.geogebra.org/m/zqkqf8nb)  


For transitive relations like above image, we can go one step at a time.

We can see above that w is function of u and v and can be represented as:  

$w = f(u)g(v)$   
 
where f(u) = u and g(v) = $\frac{1}{v}$  

using our earlier discussion:  

$\frac{\partial{w}}{\partial{x}} =  \frac{\partial{w}}{\partial{u}}\frac{\partial{u}}{\partial{x}} + \frac{\partial{w}}{\partial{v}}\frac{\partial{v}}{\partial{x}}$  

And we can find $\frac{\partial{u}}{\partial{x}}$ and $\frac{\partial{v}}{\partial{x}}$ applying the same logic (Note they are again diamond shaped functions of x).
