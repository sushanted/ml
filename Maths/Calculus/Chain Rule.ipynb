#### Chain Rule



$ \frac{dy}{dx} $ is the derivative or instantaneous slope at a given point.  

$ \frac{d}{dx} x^2  = 2x$$\tag{1.1}$  
It means when x changes by dx, y changes by 2x.dx  

$ dy = 2x dx$  



Consider following functions:  

$ y = f(x) $  
$ z = g(y) $  

If we fix the value of x to a constant c, values of y and z above will also be fixed to the constants f(c) and g(f(c)) respectively.  


When we fix x = c, $\frac{dy}{dx}$, which is also a function of x, $f'(x) =\frac{dy}{dx}$ is also fixed. For example, in (1) above, $f'(x) =\frac{dy}{dx} = 2x = 2c$$\tag{1.2}$  

$dy = f'(x)dx = f'(c)dx$  

Means for a fixed value of x=c, dy is only function of dx.  

We can have same argument for dz, for a fixed value of y, say $c_1 = f(c)$ (which we got by fixing x=c first),  

dz is only function of dy (and transitively dx because dy is function of dx)  

Now consider following figure:(https://www.geogebra.org/m/k55rxwde)  

![chain_rule](chainrule.png)

(Note that the sharing of dy(segment BD) between the two right angle triangle illustrates the chain-effect of dx on dz)
(Also note that dx,dy and dz are almost 0 (lim -> 0), but for illustration we have shown their concrete scaled values in above figure)

Here  

slope of line AB = m1 = $\frac{dy}{dx}$  

$dy = m1*dx$$\tag{1.3}$  

slope of line BC = m2 = $\frac{dz}{dy}$  

$dz = m2*dy$$\tag{1.4}$  

Note that m1 and m2 are constants(fixed) as we discussed in (1.2)  

$dz = m2*m1*dx $$\tag {from 1.3 and 1.4 above}$  

$\frac{dz}{dx} = m2*m1$  

Note segments DC and DE are exactly same in above figure, therefore   

slope of line AE = $\frac{dz}{dx} = m2*m1 = \frac{dz}{dy}\frac{dy}{dx}$  







#### An example

Now consider following relations  

$ x = w^2 $  
$ y = x^3 $  
$ z = 5y  $  

$ \frac{dx}{dw} = 2w $  
$ \frac{dy}{dx} = 3x^2 $  
$ \frac{dz}{dy} = 5 $  

$ dx = 2w dw $  
$ dy = 3x^2 dx $  
$ dz = 5 dy $  

$ dz = 5 *3x^2 dx $  
$ dz = 5 *3x^2 * 2w dw $  

$ \therefore$  

$ \frac{dz}{dx} = 5 *3x^2 = \frac{dz}{dy}\frac{dy}{dx} $  

$ \frac{dz}{dw} = 5 *3x^2 * 2w = \frac{dz}{dy}\frac{dy}{dx}\frac{dx}{dw} $  

Also  

$ \frac{dz}{dw} = \frac{dz}{dy}\frac{dy}{dw}$  


##### Important note: If we know a derivative $\frac{dy}{dw}$ and we know the relation between z and y, we can calculate $\frac{dz}{dw} $ using chain rule  

#### In General:

$ z = g(y) $  
$ y = f(x) $  

$ \frac{dz}{dy} = g'(y) $  

$ dz = g'(y) dy $$\tag {2.1}$  

$ \frac{dy}{dx} = f'(x) $  

$ dy = f'(x)dx $$\tag {2.2}$    

$ dz = g'(y)f'(x)dx $$\tag {from (2.1) and (2.2)}$    

$ \frac{dz}{dx} = g'(y)f'(x) $  

$ \frac{dz}{dx} = \frac{dz}{dy} \frac{dy}{dx} $  


The chain can be very long, if we know the derivative of last term to y, and if we know relation between x and y, we can get derivatove of last term to x.

$ \frac{dx_1}{dx_5} = \frac{dx_1}{dx_4}\frac{dx_4}{dx_5} $  

$ \frac{dx_1}{dx_4} = \frac{dx_1}{dx_3}\frac{dx_3}{dx_4} $ 

and so on...
