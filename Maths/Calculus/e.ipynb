### e is a special constant

derivative of $ a^x $  

$ \frac {dy}{dx} = \frac{d(a^x)}{dx} $  

$ = \frac{a^{(x+dx)} - a^x}{dx} $  

$ = \frac{a^x(a^{dx}-1)}{dx} $  

As dx approaches 0, for any value of a, we get a constant for $ \frac{(a^{dx}-1)}{dx} $  

e.g,   
for a = 2 , 0.693  
for a = 3 , 1.0986  
for a = 2.718, 1  

There must be a number for which the value should be 1, let's say that number is e , then  

$ \frac{(e^{dx}-1)}{dx} = 1 $$\tag 1$  

For very small value of dx,

$ \lim_{dx \to 0} \frac{(e^{dx}-1)}{dx} = 1 $  


$ \lim_{dx \to 0} {e^{dx}} - 1 = dx $  

$ \lim_{dx \to 0} {e^{dx}} =  1 + dx $  

$  {e} =  \lim_{dx \to 0}{(1 + dx)}^\frac{1}{dx} $  

### $  {e} =  \lim_{dx \to 0} {(1 + dx)}^\frac{1}{dx} $  


if dx = $ \frac{1}{n} $, then $ n \to \infty $  

### $  {e} =  \lim_{n \to \infty} {(1 + \frac{1}{n})}^n $  


We can observe that, if we put a small value say 0.0001 in above equation  


#### $ e = {(1 + 0.0001)}^\frac{1}{0.0001} = 2.7181 $  https://www.google.com/search?q=%281+%2B+0.0001%29%5E%281%2F0.0001%29



















#### Derivative of $ e^x  = $

$ = \frac{e^{(x+dx)} - e^x}{dx} $  

$ = \frac{e^x(e^{dx}-1)}{dx} $  

$ = e^x (1) $$\tag{from 1}$  

$ = e^x $  


$ \frac {d}{dx} e^x = e^x$$\tag 2$  



#### $ln$ is the log with base $e$  

$ln (e) = 1$  

$ln (e^2) = 2$  

$ln (e^x) = x$  



#### Getting derivative of $\ln (x)$  

$ y = \ln (x)$  

$ e^y = x$  

Differentiating with dx:  

$ \frac{d}{dx} e^y = \frac{d}{dx} x $  

$  e^y\frac{dy}{dx} = 1 $  

$  \frac{dy}{dx} = \frac{1}{e^y} = \frac{1}{x}$  











#### Getting derivative of $a^x$  

$ y = a^x$  

$ \ln y = \ln a^x$  

$ \ln y = x\ln a$  

Differentiating with dx:

$ \frac{d}{dx} \ln y = \frac{d}{dx} x\ln a$  

$ \frac{1}{y} \frac{dy}{dx}  = \ln a$  

$ \frac{dy}{dx}  = y\ln a$  

$ \frac{dy}{dx}  = a^x \ln a$  







#### $x^y$ equivalence  

$z = x^y$

$ln(z) = ln(x^y)$  

$ln(z) = yln(x)$  

$ e^{\ln(z)} = e^{yln(x)} $
