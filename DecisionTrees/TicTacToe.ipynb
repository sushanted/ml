#### Very tight fitting to learn all the data

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_text
import numpy as np
import pandas as pd

df = pd.read_csv("data/tic-tac-toe-endgame.csv")

columns = list(df.head())



# cleaning
df.replace({'b':0,'o':1,'x':2,'positive':1,'negative':-1},inplace=True)

results = df['V10']

df.drop(['V10'],axis=1,inplace=True)


classifier = DecisionTreeClassifier(random_state=0,max_depth=5)

classifier.fit(df,results)


print(export_text(classifier, feature_names=list(df.columns)))

# XOO
# X__
# X__
print(classifier.predict([[2,1,1,2,0,0,2,0,0]]))


```
