### Covariance

say, Z = XY, multiplication of all realizable values of X and Y, the order doesn't matter.

$ P_{XY} = P_X$ AND $P_Y =  P_Y$ AND $P_X$

$ \therefore EZ = E[XY] = E[YX] $

$ \therefore Cov(X,Y) = Cov(Y,X) $  

Note that  
$ Cov(X,Y) = E[(X-EX)(Y-EY)] $  

X can have m realizable values and Y can have n realizable values, so XY can have m*n realizable values, but $P_{XY}$ could be defined only for some X and Y and coulds be 0 for all other X and Y. For example we can have a graph (monthly sales report for a year) where there is one-to-one mapping between X and Y values and thus $P_{XY}$ would be defined for only m (or n) values, and is 0 for all other possible pairs.  

### Correlation

$ (a-b)^2 = a^2+b^2+2ab $  

being a square $(a-b)^2 >= 0 $  

$ a^2+b^2+2ab >= 0 $  

$ \frac{a^2+b^2}{2} - ab >= 0 $  

$ \frac{a^2+b^2}{2} >= ab $   

For 2 random variables, for any realized values of X and Y,  

$ \frac{X^2+Y^2}{2} >= XY $  

Taking expectations at both sides:


$ E[\frac{X^2+Y^2}{2}] >= E[XY] $  

$ \frac{E[X^2]+E[Y^2]}{2} >= E[XY] $ $\tag 1$    

For standardized variables U and V,  

$ U = \frac{X - EX}{\sigma_X} $  

$ V = \frac{Y - EY}{\sigma_Y} $  

$ E[U^2] = \frac{E[(X-EX)^2]}{\sigma_X^2} = \frac{Var[X]}{Var[X]} = 1$ $ \tag 2$

$ E[V^2] = \frac{E[(Y-EY)^2]}{\sigma_Y^2} = \frac{Var[Y]}{Var[Y]} = 1$ $ \tag 3$

$ E[UV] = \frac{E[(X-EX)(Y-EY)]}{\sigma_X \sigma_Y} = \frac{Cov(X,Y)}{\sigma_X \sigma_Y} = \sigma_{XY}$ $ \tag 4$

Applying (1), on U and V:

$ \frac{E[U^2]+E[V^2]}{2} >= E[UV] $ 

$ \frac{1+1}{2} >= \sigma_{XY} $  $ \tag { by 2,3,4} $

$ \sigma_{XY} <= \frac{1+1}{2}  $  

$ \sigma_{XY} <= 1  $
