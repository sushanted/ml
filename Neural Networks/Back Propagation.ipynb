

#### Over all Cost function

$ C = \frac{1}{2} \|y-a^L\|^2 $$\tag{y and $a^L$ are vectors}$

$ = \frac{1}{2} {\sqrt{(y_1-a_1^L)^2 + (y_2-a_2^L)^2 + (y_3-a_3^L)^2 + ...}}^2 $$\tag{magnitude of a vector}$  

$ = \frac{1}{2} [(y_1-a_1^L)^2 + (y_2-a_2^L)^2 + (y_3-a_3^L)^2 + ...] $

$ = \frac{1}{2} \sum_j (y_j-a^L_j)^2 $

#### Cost function for a single sample

Cost function is a function of all w and b (and any intermediate entities z,a), All these partial derivatives are part of the gradient vector $ \nabla $. A step in the opposite direction of this vector is suppose to reduce the overall cost.

$ C = \frac{1}{2} \sum_j(y_j-a^L_j)^2 $  

$ \therefore \partial C / \partial a^L_j = (a_j^L-y_j)$ $\tag 1$  

$ \because \frac{\partial}{\partial x} [(a-x)^2 + (b-y)^2] $  

$=  \frac{\partial}{\partial x} [(a-x)^2] $  

$= \frac{\partial}{\partial x} (a^2-2ax+x^2)$  

$ = 2(x-a)$

Only one j remains, because it is partial derivative  

#### BP1

j is index in the last layer  

$ \frac{\partial C}{\partial z_j^L} = \frac{\partial C}{\partial a_j^L} \frac{\partial a_j^L}{\partial z_j^L}$

$ a^L_j = \sigma(z^L_j) $ 

$ \frac{\partial C}{\partial z_j^L} = \frac{\partial C}{\partial a_j^L} \frac{\partial }{\partial z_j^L} \sigma (z_j^L)$

$ = \frac{\partial C}{\partial a_j^L} \sigma'(z_j^L)$

Follows a good source to understand BP : https://www.jeremyjordan.me/neural-networks-training/

**Alternate proof by expansion (without chain rule):**

$ C = \frac{1}{2}\sum_k(y_k-a_k^{L})^2 $  

$  = \frac{1}{2}(y_1-a_1^{L})^2 + \frac{1}{2}(y_2-a_2^{L})^2 $  

$ \frac{\partial C}{\partial z_1^L} = (a_1^L-y_1) \frac{\partial }{\partial z_1^L} a_1^L + (a_2^L-y_2) \frac{\partial }{\partial z_1^L} a_2^L $  

$  = (a_1^L-y_1) \frac{\partial }{\partial z_1^L} a_1^L + 0 $  

$  = (a_1^L-y_1) \frac{\partial }{\partial z_1^L} \sigma(z_1^L)  $  

$  = (a_1^L-y_1) \sigma'(z_1^L)  $$\tag{2}$  

$  =  \frac{\partial C}{\partial a^L_1} \sigma'(z_1^L)  $$\tag{from (1)}$

#### BP2

Consider following setup:  
3 neurons in layer l and 2 neurons in layer l+1  
layer l+1 is the final layer  
j is the index in l layer  
k is the index in l+1 layer  

consider the first(j=1) neuron in layer l, it is feeding into both the neurons (k=1 and 2) of layer l+1  


$ C = \frac{1}{2}\sum_k(y_k-a_k^{l+1})^2 $  

$  = \frac{1}{2}(y_1-a_1^{l+1})^2 + \frac{1}{2}(y_2-a_2^{l+1})^2 $  

$ \frac{\partial C}{\partial z_1^l} = -(y_1-a_1^{l+1})\frac{\partial{a_1^{l+1}}}{\partial{z_1^l}} + -(y_2-a_2^{l+1})\frac{\partial{a_2^{l+1}}}{\partial{z_1^l}} $  

$ \frac{\partial C}{\partial z_1^l} = (a_1^{l+1}-y_1)\frac{\partial{a_1^{l+1}}}{\partial{z_1^l}} + (a_2^{l+1}-y_1)\frac{\partial{a_2^{l+1}}}{\partial{z_1^l}} $  

Applying chain rule:  

$ \frac{\partial C}{\partial z_1^l} = (a_1^{l+1}-y_1)\frac{\partial{a_1^{l+1}}}{\partial{z_1^l}} + (a_2^{l+1}-y_1)\frac{\partial{a_2^{l+1}}}{\partial{z_1^l}} $  

$ \frac{\partial C}{\partial z_1^l} = (a_1^{l+1}-y_1)\frac{\partial{a_1^{l+1}}}{\partial{z_1^{l+1}}}\frac{\partial{z_1^{l+1}}}{\partial{z_1^l}} + (a_2^{l+1}-y_2)\frac{\partial{a_2^{l+1}}}{\partial{z_2^{l+1}}}\frac{\partial{z_2^{l+1}}}{\partial{z_1^l}} $ $\tag 3$    

Note the indices carefully above for $\partial z^{l+1}$, $a_1$ is function of $z_1$ while $a_2$ is function of $z_2$  

Note below that we are considering only $ a_1^l $ because it is only function of $ z_1^l $ ,  $ a_2^l, a_3^l $ would become zero in differentiation with $ z_1^l $.

$ z_1^{l+1} = w_{11}^{l+1}a_1^l+b_1^{l+1} = w_{11}^{l+1}\sigma(z_1^l)+b_1^{l+1}$ and  

$ z_2^{l+1} = w_{21}^{l+1}a_1^l+b_2^{l+1} = w_{21}^{l+1}\sigma(z_1^l)+b_2^{l+1} $  

therefore differentiating with $\partial z_1^l$  

$ \frac{\partial{z_1^{l+1}}}{\partial{z_1^l}} = w_{11}^{l+1}\sigma'(z_1^l) $ and  

$ \frac{\partial{z_2^{l+1}}}{\partial{z_1^l}} = w_{21}^{l+1}\sigma'(z_1^l) $  

Substituting in (3)  

$ \frac{\partial C}{\partial z_1^l} = (a_1^{l+1}-y_1)\frac{\partial{a_1^{l+1}}}{\partial{z_1^{l+1}}}w_{11}^{l+1}\sigma'(z_1^l) + (a_2^{l+1}-y_2)\frac{\partial{a_2^{l+1}}}{\partial{z_2^{l+1}}}w_{21}^{l+1}\sigma'(z_1^l) $$\tag 4$  


from (1),  
$(a_1^{l+1}-y_1) = \frac{\partial C}{\partial{a_1^{l+1}}}$ and   
$(a_2^{l+1}-y_2) = \frac{\partial C}{\partial{a_2^{l+1}}}$  

Substituting in (4)  

$ \frac{\partial C}{\partial z_1^l} = \frac{\partial C}{\partial{a_1^{l+1}}}\frac{\partial{a_1^{l+1}}}{\partial{z_1^{l+1}}}w_{11}^{l+1}\sigma'(z_1^l) + \frac{\partial C}{\partial{a_2^{l+1}}}\frac{\partial{a_2^{l+1}}}{\partial{z_2^{l+1}}}w_{21}^{l+1}\sigma'(z_1^l) $   

$ \frac{\partial C}{\partial z_1^l} = \frac{\partial C}{\partial{z_1^{l+1}}}w_{11}^{l+1}\sigma'(z_1^l) + \frac{\partial C}{\partial{z_2^{l+1}}}w_{21}^{l+1}\sigma'(z_1^l) $  


**Alternate proof by expansion (without the chain rule):**  

$ C = \frac{1}{2}\sum_k(y_k-a_k^{l+1})^2 $  

$  = \frac{1}{2}(y_1-a_1^{l+1})^2 + \frac{1}{2}(y_2-a_2^{l+1})^2 $  

$ \frac{\partial C}{\partial z_1^l} = -(y_1-a_1^{l+1})\frac{\partial{a_1^{l+1}}}{\partial{z_1^l}} + -(y_2-a_2^{l+1})\frac{\partial{a_2^{l+1}}}{\partial{z_1^l}} $  

$ \frac{\partial C}{\partial z_1^l} = (a_1^{l+1}-y_1)\frac{\partial{a_1^{l+1}}}{\partial{z_1^l}} + (a_2^{l+1}-y_2)\frac{\partial{a_2^{l+1}}}{\partial{z_1^l}} $  

$  = (a_1^{l+1}-y_1)\frac{\partial}{\partial{z_1^l}} \sigma ({z_1^{l+1}}) + (a_2^{l+1}-y_2)\frac{\partial}{\partial{z_1^l}} \sigma ({z_2^{l+1}}) $  

$  = (a_1^{l+1}-y_1)\sigma'({z_1^{l+1}})\frac{\partial}{\partial{z_1^l}} {z_1^{l+1}} + (a_2^{l+1}-y_2)\sigma'({z_2^{l+1}})\frac{\partial}{\partial{z_1^l}}{z_2^{l+1}}  $$\tag{5}$  

$ \frac{\partial}{\partial{z_1^l}} {z_1^{l+1}} = \frac{\partial}{\partial{z_1^l}} (w^{l+1}_{11}a^l_1 + w^{l+1}_{12}a^l_2 + w^{l+1}_{13}a^l_3 + b^{l+1}_1) $  

$ = \frac{\partial}{\partial{z_1^l}} (w^{l+1}_{11}a^l_1) + 0 + 0 + 0 $$\tag{$a^l_2,a^l_3 $ and $b^{l+1}_1$ are not function of $z^l_1$}$  

$ = w^{l+1}_{11} \frac{\partial}{\partial{z_1^l}} \sigma(z^l_1) $  

$ = w^{l+1}_{11} \sigma'(z^l_1) $$\tag{6}$  

$ \frac{\partial}{\partial{z_1^l}} {z_2^{l+1}} = \frac{\partial}{\partial{z_1^l}} (w^{l+1}_{21}a^l_1 + w^{l+1}_{22}a^l_2 + w^{l+1}_{23}a^l_3 + b^{l+1}_2) $  

$ = \frac{\partial}{\partial{z_1^l}} (w^{l+1}_{21}a^l_1) + 0 + 0 + 0 $$\tag{$a^l_2,a^l_3 $ and $b^{l+1}_2$ are not function of $z^l_1$}$  

$ = w^{l+1}_{21} \frac{\partial}{\partial{z_1^l}} \sigma(z^l_1) $  

$ = w^{l+1}_{21} \sigma'(z^l_1) $$\tag{7}$  

Substituting (6) and (7) in (5)  

$ \frac{\partial C}{\partial z_1^l} = (a_1^{l+1}-y_1)\sigma'({z_1^{l+1}}) w^{l+1}_{11} \sigma'(z^l_1) + (a_2^{l+1}-y_2)\sigma'({z_2^{l+1}}) w^{l+1}_{21} \sigma'(z^l_1) $  

$ = \frac{\partial C}{\partial z_1^{l+1}} w^{l+1}_{11} \sigma'(z^l_1) + \frac{\partial C}{\partial z_2^{l+1}} w^{l+1}_{21} \sigma'(z^l_1) $

**Generalizations:**  

$ \frac{\partial C}{\partial z_1^l} = \sum_k\frac{\partial C}{\partial{z_k^{l+1}}}w_{k1}^{l+1}\sigma'(z_1^l)  $  

And generically, for any $j^{th}$ neuron in layer l:  

$ \frac{\partial C}{\partial z_j^l} = \sum_k\frac{\partial C}{\partial{z_k^{l+1}}}w_{kj}^{l+1}\sigma'(z_j^l)  $  


Note: BP2 can be used iteratively for each layer L-1 to 1

### BP3 and BP4

Refer Maths/Chain Rule.ipynb

if we know $ \frac{\partial{C}}{\partial z_j^l}$ in any layer l and for any $j^{th}$ neuron, we can easily calculate the BP3 and BP4 for that neuron, (bias and weights respectively).

#### BP3

$\frac{\partial C}{\partial b_j^l} = \frac{\partial C}{\partial z_j^l}\frac{\partial z_j^l}{\partial b_j^l}$ $\tag 8$

$ z_j^l = \sum_i w_{ji}^l a_i^{l-1} + b_j^l $ where i is the index in the previous layer  

Differentiating with $b_j^l$  

$ \frac{\partial z_j^l}{\partial b_j^l} = 0 + 1 $  

Substituting in (8)

$ \frac{\partial C}{\partial b_j^l}  = \frac{\partial C}{\partial z_j^l} (1) $  

$  = \frac{\partial C}{\partial z_j^l}  $  

#### BP4

$\frac{\partial C}{\partial w_{ji}^l} = \frac{\partial C}{\partial z_j^l}\frac{\partial z_j^l}{\partial w_{ji}^l}$ $\tag 9$  

$ z_j^l = \sum_i w_{ji}^l a_i^{l-1} + b_j^l $ where i is the index in the previous layer  

$  =  w_{j1}^l a_1^{l-1}+w_{j2}^l a_2^{l-1}+ .... + b_j^l $ 

Differentiating with $ w_{j1}^l $,

$ \frac{\partial z_j^l}{\partial w_{j1}^l} = a_1^{l-1} + 0 + 0+ ... +0 $  

similarly:  

$ \frac{\partial z_j^l}{\partial w_{j2}^l} = 0 + a_2^{l-1} + 0+ ... +0 $

In general for any i

$ \frac{\partial z_j^l}{\partial w_{ji}^l} =  a_i^{l-1} $

$ \frac{\partial C}{\partial w_{ji}^l} = \frac{\partial C}{\partial z_j^l} a_i^{l-1} $
